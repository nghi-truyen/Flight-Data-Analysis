```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## SN_list data

### Data set 

```python
SN_list = pd.read_excel(r'C:\Users\Nghi Truyen\Desktop\Projet_Recherche_5A\Flight_Data_Analysis\INSA-Airbus FHS - Data set_bis.xlsx', sheet_name='SN list')
SN_list.info()
SN_list.head()
```

```python
PN = "Part Number"
SN = "Serial Number"
Description = "Description"
Status = "Current SN Status Description"
Company = "Company"
TSI = "Hour ageing Since Installation"
TSN = "Hour ageing Since New"
SND = "Since New Date"
```

### Data visualization 

```python
plt.hist(SN_list[TSI],bins=20)
plt.ylabel('Frequency')
plt.xlabel('Hour ageing since installation')
plt.title('Time since installation for 3 types A, B and C')
plt.figure()
plt.hist(SN_list[TSN],bins=20)
plt.ylabel('Frequency')
plt.xlabel('Hour ageing since new')
plt.title('Time since new for 3 types A, B and C')

print("Number of samples in total : ", len(SN_list[TSI]))
```
